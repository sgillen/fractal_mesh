```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def plot_mean_std(x_data, data):
    means = np.array(data.mean(axis=0))
    stds = np.array(data.std(axis=0))
    plt.plot(x_data, means)
    plt.fill_between(x_data, (means-stds), (means+stds), alpha=.5)
    
def plot_mean_minmax(x_data, data):
    means = np.array(data.mean(axis=0))
    mins = np.array(data.min(axis=0))
    maxs = np.array(data.max(axis=0))
    
    plt.plot(x_data, means)
    plt.fill_between(x_data, (mins), (maxs), alpha=.5)
    
deltas = [.4,.3,.2,.1]
```

```python
ms_i = pd.read_csv('mesh_sizes/iden_25.csv', index_col=0)
ms_in = pd.read_csv('mesh_sizes/iden_noise1_25.csv', index_col=0)

ms_m = pd.read_csv('mesh_sizes/mdim_25.csv', index_col=0)
ms_mn = pd.read_csv('mesh_sizes/mdim_noise1_25.csv', index_col=0)

ms = [ms_i, ms_in, ms_m, ms_mn]
```

```python
for m in ms:
    print(m.mean(axis=0))

print('\n\n\n')
    
for m in ms:
    print(m.std(axis=0))
```

```python
for m in ms:
    plt.plot(deltas[0:3], m.mean(axis=0)[0:3])
    
plt.figure()


for m in ms:
    plot_mean_minmax(deltas , m)

plt.figure()

for m in ms:
    plot_mean_std(deltas , m)
```
